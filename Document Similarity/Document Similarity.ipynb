```python
import math
import pandas as pd

with open('document3.txt', 'r') as file:
    sentences = file.read().split(',')
```

```python

# Create document vectors for the sentences
document_vectors = []

# tokenize a sentence and create a document vector
def create_document_vector(sentence, vocabulary):
    words = sentence.lower().split()
    document_vector = [words.count(word) for word in vocabulary]
    return document_vector
```

```python

# Function to tokenize a sentence into words and remove stopwords
def tokenize(sentence):
    words = sentence.lower().split()
    return [word for word in words if word not in stop_words]

# Tokenize the sentences into words and remove stopwords
words = [tokenize(sentence) for sentence in sentences]

# Create the vocabulary based on all unique words in the sentences
vocabulary = list(set(word for sentence_words in words for word in sentence_words if word not in stop_words))
```

```python

# Create document vectors for each sentence
for sentence in sentences:
    document_vector = create_document_vector(sentence, vocabulary)
    document_vectors.append(document_vector)
```

```python

# calculate cosine similarity with a check for division by zero
def cosine_similarity(x, y):
    if all(v == 0 for v in x) or all(v == 0 for v in y):
        return 0.0  # Set similarity to zero when either vector has all zeros
    prod = sum(x[i] * y[i] for i in range(len(x)))
    modx = math.sqrt(sum(i**2 for i in x))
    mody = math.sqrt(sum(i**2 for i in y))
#     print(modx, mody, prod)
    return prod / (modx * mody)

```

```python
# Initialize a similarity matrix
similarity_matrix = []

# Calculate and store the cosine similarity between each pair of sentences
for i in range(len(sentences)):
    row = []
    for j in range(len(sentences)):
        if i == j:
            similarity = 1.0  # The similarity of a sentence with itself is 1.0
        else:
            x = document_vectors[i]
            y = document_vectors[j]
            similarity = cosine_similarity(x, y)
        row.append(similarity)
    similarity_matrix.append(row)

```

```python
# Create a Pandas DataFrame for the similarity matrix
df = pd.DataFrame(similarity_matrix, columns=[f"Sentence {i + 1}" for i in range(len(sentences))], index=[f"Sentence {i + 1}" for i in range(len(sentences))])

# Display the DataFrame
print(df)
```
